**Данный файл предназначен для парсинга данных с сайта http://mega-mult.ru/**

***Импорты***

```python
import requests
from multiprocessing.dummy import Pool as ThreadPool
import time
from bs4 import BeautifulSoup
from functools import partial
```

***Основной код***

```python
st_accept = "text/html"
st_useragent = "Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.10136"

headers = {
    "Accept": st_accept,
    "User-Agent": st_useragent
}
```

```python
pools = ThreadPool(5)
urls = [f"http://mega-mult.ru/multfilm/page/{i}" for i in range(17)]
start_time1 = time.time()
req_with_headers = partial(requests.get, params=headers)
results = pools.map(req_with_headers, urls)
for res in results:
    res.encoding = "Windows-1251"
start_time2 = time.time()
print(start_time2-start_time1)
bs_with_lxml = partial(BeautifulSoup, features="lxml")
bs_results = pools.map(bs_with_lxml, [res.text for res in results])
start_time3 = time.time()
print(start_time3-start_time2)
```
